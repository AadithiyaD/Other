```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,],[3,-1,55], label='EAGAEDFAS', color='red', linewidth =2, linestyle='--', marker='X', markersize=12,markeredgecolor="Black")


plt.title("GRAPH", fontdict={"fontsize":50})
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")

plt.xticks([0,1,2,4])

plt.legend()

#plt.savefig("GRAPHHHHH",dpi=300)


plt.show()
```

```python
x=np.linspace(0,10,5)
y=np.linspace(0,10,5)

plt.plot(x,y, 'r.:', label='[color][marker][line style]')
plt.plot(x,x**2)

plt.legend()

plt.show()

plt.plot(x,x**2)
plt.show()
```
